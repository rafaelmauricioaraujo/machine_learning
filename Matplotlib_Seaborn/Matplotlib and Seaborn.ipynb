# Matplotlib and Seaborn
### Bivariate Visualizations - Visualization of two variables 

01 - Scatterplots for quantitative variables vs quantitative variables

02 - Violinplots for quantitative variables vs qualitative variables

03 - Clustered bar charts for qualitative variables vs qualitative variables

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

wine_data = pd.read_csv('winequality_edited.csv')
print(wine_data.shape)
wine_data.head()
```

```python
plt.scatter(data = wine_data, y='pH', x='fixed_acidity');
plt.xlabel('Fixed Acidity')
plt.ylabel('PH')
```

```python
sb.regplot(data = wine_data, y='pH', x='fixed_acidity');
plt.xlabel('Fixed Acidity')
plt.ylabel('PH')
```

```python
def log_trans(x, inverse = False):
    if not inverse:
        return np.log10(x)
    else:
        return np.power(10, x)

sb.regplot(wine_data['pH'], wine_data['fixed_acidity'].apply(log_trans))
tick_locs = [5.0, 5.5, 6.0, 6.5, 7, 7.5]
plt.yticks(log_trans(tick_locs), tick_locs)
```

# Overplotting:
### Resolve with:

01 - Sampling

02 - Transparency

03 - Jitter

```python
sb.regplot(data = wine_data, x='quality', y='fixed_acidity');
plt.xlabel('Quality')
plt.ylabel('Fixed Acidity')
```

```python
sb.regplot(data = wine_data, x='quality', y='fixed_acidity', x_jitter = 0.3);
plt.xlabel('Quality')
plt.ylabel('Fixed Acidity')
```

```python
# Add some tranparency
sb.regplot(data = wine_data, x='quality', y='fixed_acidity', x_jitter = 0.3, scatter_kws = {'alpha' : 1/5} );
plt.xlabel('Quality')
plt.ylabel('Fixed Acidity')
```
